```python
from scipy.stats import chi2_contingency

def chi_square_test(data, feature, target):

    # contingency table
    table = pd.crosstab(data[feature], data[target])

    # chi-square test
    chi2, p, dof, expected = chi2_contingency(table)

    print(f"\nFeature: {feature}")
    print(f"Chi-square statistic: {chi2:.4f}")
    print(f"P-value: {p:.6f}")

    if p < 0.05:
        print("SIGNIFICANT relationship with target")
    else:
        print("NOT significant relationship with target")

    return table
```

```python
from scipy.stats import ttest_ind
import pandas as pd
import numpy as np

TARGET = "target"

# ALL MODEL
all_model_df = df.copy()

# FUNDED MODEL
funded_model_df = df[df["funded_ind"] == 1].copy()


def significance_level(p):

    if p < 0.001:
        return "*** Highly Significant"

    elif p < 0.01:
        return "** Very Significant"

    elif p < 0.05:
        return "* Significant"

    else:
        return "NS Not Significant"


def run_ttest(data):

    # numeric columns
    num_cols = data.select_dtypes(
        include=["int64", "float64"]
    ).columns.tolist()

    # remove target
    if TARGET in num_cols:
        num_cols.remove(TARGET)

    results = []

    for col in num_cols:

        try:

            temp = data[[col, TARGET]].replace(
                [np.inf, -np.inf],
                np.nan
            )

            # split groups
            grp0 = temp[temp[TARGET] == 0][col].dropna()
            grp1 = temp[temp[TARGET] == 1][col].dropna()

            # skip if insufficient data
            if len(grp0) < 2 or len(grp1) < 2:
                continue

            # Welch t-test
            t_stat, p_value = ttest_ind(
                grp0,
                grp1,
                equal_var=False
            )

            results.append({
                "feature": col,
                "mean_target_0": grp0.mean(),
                "mean_target_1": grp1.mean(),
                "mean_diff": grp1.mean() - grp0.mean(),
                "t_statistic": t_stat,
                "p_value": p_value,
                "target_0_count": len(grp0),
                "target_1_count": len(grp1),
                "significance_level": significance_level(p_value)
            })

        except Exception as e:
            print(f"Skipping {col}: {e}")

    results_df = pd.DataFrame(results)

    return results_df.sort_values(
        ["p_value", "t_statistic"],
        ascending=[True, False]
    )


# RUN FOR BOTH MODELS
ttest_all = run_ttest(all_model_df)

ttest_funded = run_ttest(funded_model_df)

print("ALL MODEL")
display(ttest_all.head(20))

print("FUNDED MODEL")
display(ttest_funded.head(20))
```

```python
from scipy.stats import chi2_contingency
import pandas as pd
import numpy as np

TARGET = "target"

# ALL MODEL
all_model_df = df.copy()

# FUNDED MODEL
funded_model_df = df[df["funded_ind"] == 1].copy()


def significance_level(p):

    if p < 0.001:
        return "*** Highly Significant"

    elif p < 0.01:
        return "** Very Significant"

    elif p < 0.05:
        return "* Significant"

    else:
        return "NS Not Significant"


def run_chi_square(data, max_unique=50):

    # categorical columns
    cat_cols = data.select_dtypes(
        include=["object", "category", "bool"]
    ).columns.tolist()

    results = []

    for col in cat_cols:

        try:

            temp = data[[col, TARGET]].copy()

            # handle missing
            temp[col] = temp[col].astype("object").fillna("MISSING")

            # skip high cardinality
            if temp[col].nunique() > max_unique:
                continue

            # contingency table
            table = pd.crosstab(temp[col], temp[TARGET])

            # chi-square test
            chi2, p_value, dof, expected = chi2_contingency(table)

            # overall bad rates
            bad_rate = temp.groupby(col)[TARGET].mean()

            results.append({
                "feature": col,
                "num_categories": temp[col].nunique(),
                "chi_square": chi2,
                "degrees_of_freedom": dof,
                "p_value": p_value,
                "overall_bad_rate": temp[TARGET].mean(),
                "max_bad_rate": bad_rate.max(),
                "min_bad_rate": bad_rate.min(),
                "bad_rate_range": bad_rate.max() - bad_rate.min(),
                "significance_level": significance_level(p_value)
            })

        except Exception as e:
            print(f"Skipping {col}: {e}")

    results_df = pd.DataFrame(results)

    return results_df.sort_values(
        ["p_value", "chi_square"],
        ascending=[True, False]
    )


# RUN FOR BOTH MODELS
chi_all = run_chi_square(all_model_df)

chi_funded = run_chi_square(funded_model_df)

print("ALL MODEL")
display(chi_all.head(20))

print("FUNDED MODEL")
display(chi_funded.head(20))
```
