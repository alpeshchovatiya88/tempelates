```python
def categorical_distribution_summary(data, features, model_name):
    rows = []

    for col in features:
        temp = (
            data[col]
            .astype("object")
            .fillna("MISSING")
            .value_counts(dropna=False)
            .reset_index()
        )

        temp.columns = ["level", "count"]
        temp["feature"] = col
        temp["model"] = model_name
        temp["pct"] = temp["count"] / len(data)

        rows.append(temp)

    return pd.concat(rows, ignore_index=True)


def compare_distributions(datasets, features):
    """
    datasets = {
        "All_model": all_model,
        "Funded_model": funded_model,
        ...
    }
    """

    dist = pd.concat(
        [
            categorical_distribution_summary(df, features, model_name)
            for model_name, df in datasets.items()
        ],
        ignore_index=True,
    )

    result = (
        dist.pivot_table(
            index=["feature", "level"],
            columns="model",
            values=["count", "pct"],
            aggfunc="first",
        )
        .reset_index()
    )

    result.columns = [
        "_".join([x for x in col if x]).strip("_")
        if isinstance(col, tuple)
        else col
        for col in result.columns
    ]

    return result
```

```python
result = compare_distributions(
    datasets={
        "All_model": all_model,
        "Funded_model": funded_model
    },
    features=categorical_cols
)

display(result)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_stacked_categorical(
    datasets,
    categorical_cols,
    normalize=True,
    top_n=None,
    missing_label="MISSING",
    figsize=(10, 5),
    rotation=45
):
    """
    datasets example:
        {
            "All Model": all_model,
            "Funded Model": funded_model
        }

    categorical_cols example:
        ["CRD_SCORE_bin", "STATE", "CHANNEL"]
    """

    for col in categorical_cols:
        rows = []

        for model_name, df in datasets.items():
            temp = (
                df[col]
                .astype("object")
                .fillna(missing_label)
                .value_counts(dropna=False)
                .reset_index()
            )

            temp.columns = ["level", "count"]
            temp["model"] = model_name
            rows.append(temp)

        plot_df = pd.concat(rows, ignore_index=True)

        if top_n is not None:
            top_levels = (
                plot_df.groupby("level")["count"]
                .sum()
                .sort_values(ascending=False)
                .head(top_n)
                .index
            )
            plot_df = plot_df[plot_df["level"].isin(top_levels)]

        pivot_df = plot_df.pivot_table(
            index="model",
            columns="level",
            values="count",
            aggfunc="sum",
            fill_value=0
        )

        if normalize:
            pivot_df = pivot_df.div(pivot_df.sum(axis=1), axis=0) * 100
            ylabel = "Percent"
        else:
            ylabel = "Count"

        ax = pivot_df.plot(
            kind="bar",
            stacked=True,
            figsize=figsize
        )

        ax.set_title(f"Stacked Distribution - {col}")
        ax.set_xlabel("Model")
        ax.set_ylabel(ylabel)
        ax.tick_params(axis="x", rotation=rotation)

        for container in ax.containers:
            ax.bar_label(
                container,
                fmt="%.1f" if normalize else "%.0f",
                label_type="center",
                fontsize=8
            )

        ax.legend(
            title=col,
            bbox_to_anchor=(1.05, 1),
            loc="upper left"
        )

        plt.tight_layout()
        plt.show()
```

```python
plot_stacked_categorical(
    datasets={
        "All Model": all_model,
        "Funded Model": funded_model
    },
    categorical_cols=categorical_cols,
    normalize=True,
    top_n=10
)
```
