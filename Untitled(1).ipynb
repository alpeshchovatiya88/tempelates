```python
# =========================
# 4. Overall data distribution summary
# =========================

def dataset_summary(data, name):
    return pd.DataFrame({
        "model": name,
        "rows": [len(data)],
        "target_bad_count": [(data["target"] == 1).sum()],
        "target_good_count": [(data["target"] == 0).sum()],
        "target_rate": [data["target"].mean()],
        "missing_target": [data["target"].isna().sum()]
    })

summary = pd.concat([
    dataset_summary(all_model, "all_model"),
    dataset_summary(funded_model, "funded_model")
])

summary
```

```python
# =========================
# 5. Numeric feature distribution comparison
# =========================

def numeric_distribution(data, features, model_name):
    rows = []

    for col in features:
        s = data[col]

        rows.append({
            "model": model_name,
            "feature": col,
            "count": s.notna().sum(),
            "missing": s.isna().sum(),
            "missing_rate": s.isna().mean(),
            "mean": s.mean(),
            "std": s.std(),
            "min": s.min(),
            "p1": s.quantile(0.01),
            "p5": s.quantile(0.05),
            "p25": s.quantile(0.25),
            "median": s.quantile(0.50),
            "p75": s.quantile(0.75),
            "p95": s.quantile(0.95),
            "p99": s.quantile(0.99),
            "max": s.max()
        })

    return pd.DataFrame(rows)

num_dist = pd.concat([
    numeric_distribution(all_model, numeric_features, "all_model"),
    numeric_distribution(funded_model, numeric_features, "funded_model")
])

num_dist.head()
```

```python
# =========================
# 6. Categorical feature distribution comparison
# =========================

def categorical_distribution(data, features, model_name, top_n=20):
    rows = []

    for col in features:
        temp = (
            data[col]
            .astype("object")
            .fillna("MISSING")
            .value_counts(dropna=False)
            .head(top_n)
            .reset_index()
        )

        temp.columns = ["level", "count"]
        temp["feature"] = col
        temp["model"] = model_name
        temp["percent"] = temp["count"] / len(data)

        rows.append(temp)

    return pd.concat(rows, ignore_index=True)

cat_dist = pd.concat([
    categorical_distribution(all_model, categorical_features, "all_model"),
    categorical_distribution(funded_model, categorical_features, "funded_model")
])

cat_dist.head()
```

```python
# =========================
# 7. Relationship: numeric features vs target
# =========================

def numeric_target_relationship(data, features, model_name, bins=10):
    rows = []

    for col in features:
        temp = data[[col, "target"]].copy()
        temp = temp[temp["target"].notna()]

        if temp[col].notna().sum() < 100:
            continue

        try:
            temp["bin"] = pd.qcut(temp[col], q=bins, duplicates="drop")
        except Exception:
            continue

        grouped = temp.groupby("bin", observed=False).agg(
            count=("target", "size"),
            bad_count=("target", "sum"),
            bad_rate=("target", "mean"),
            min_value=(col, "min"),
            max_value=(col, "max"),
            mean_value=(col, "mean")
        ).reset_index()

        grouped["feature"] = col
        grouped["model"] = model_name

        rows.append(grouped)

    return pd.concat(rows, ignore_index=True)

num_target_rel = pd.concat([
    numeric_target_relationship(all_model, numeric_features, "all_model"),
    numeric_target_relationship(funded_model, numeric_features, "funded_model")
])

num_target_rel.head()
```

```python
# =========================
# 8. Relationship: categorical features vs target
# =========================

def categorical_target_relationship(data, features, model_name, min_count=100):
    rows = []

    for col in features:
        temp = data[[col, "target"]].copy()
        temp[col] = temp[col].astype("object").fillna("MISSING")

        grouped = temp.groupby(col).agg(
            count=("target", "size"),
            bad_count=("target", "sum"),
            bad_rate=("target", "mean")
        ).reset_index()

        grouped = grouped[grouped["count"] >= min_count]
        grouped["feature"] = col
        grouped["model"] = model_name
        grouped = grouped.rename(columns={col: "level"})

        rows.append(grouped)

    return pd.concat(rows, ignore_index=True)

cat_target_rel = pd.concat([
    categorical_target_relationship(all_model, categorical_features, "all_model"),
    categorical_target_relationship(funded_model, categorical_features, "funded_model")
])

cat_target_rel.head()
```

2. Region distribution

```python
def region_distribution(data, model_name, region_col="PRK_PROVINCE"):

    temp = (
        data[region_col]
        .fillna("MISSING")
        .value_counts(dropna=False)
        .reset_index()
    )

    temp.columns = ["region", "count"]

    temp["percent"] = (
        temp["count"] / temp["count"].sum()
    ).round(4)

    temp["model"] = model_name

    return temp


all_region = region_distribution(all_model, "all_model")

funded_region = region_distribution(funded_model, "funded_model")

region_compare = pd.concat(
    [all_region, funded_region],
    ignore_index=True
)

region_compare
```

3. Side-by-side comparison table

```python
all_region_pivot = (
    all_region[["region", "percent"]]
    .rename(columns={"percent": "all_model_pct"})
)

funded_region_pivot = (
    funded_region[["region", "percent"]]
    .rename(columns={"percent": "funded_model_pct"})
)

region_final = (
    all_region_pivot
    .merge(
        funded_region_pivot,
        on="region",
        how="outer"
    )
)

region_final["pct_diff"] = (
    region_final["funded_model_pct"]
    - region_final["all_model_pct"]
).round(4)

region_final = region_final.sort_values(
    "all_model_pct",
    ascending=False
)

region_final
```

5. Plot comparison

```python
def region_bad_rate(data, model_name, region_col="PRK_PROVINCE"):

    temp = (
        data
        .groupby(region_col)
        .agg(
            total=("target", "size"),
            bads=("target", "sum"),
            bad_rate=("target", "mean")
        )
        .reset_index()
    )

    temp["model"] = model_name

    return temp


all_bad = region_bad_rate(all_model, "all_model")

funded_bad = region_bad_rate(funded_model, "funded_model")

region_bad_final = pd.concat(
    [all_bad, funded_bad],
    ignore_index=True
)

region_bad_final
```
