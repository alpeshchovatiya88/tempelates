```python
# ============================================================
# 0. IMPORTS + SETTINGS
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import ttest_ind, chi2_contingency
from sklearn.metrics import roc_auc_score

pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 200)

sns.set_theme(style="whitegrid")


# ============================================================
# 1. LOAD DATA
# ============================================================

features = [
    "APPLICATION_NUMBER",
    "CRD_SCORE_bin",
    "CRD_SCORE",
    "CRD_SCORE_max",
    "CRD_MONTHS_TU",
    "CRD_OPEN_TRADE_TU",
    "CRD_PUBLIC_DEROG_TU",
    "CRA_COLLECTION",
    "CRA_JUDGEMENT",
    "CRA_BANKRUPTCY_FORECLOSURE_IND",
    "ECON_REGION",
    "REG_REGION_CODE",
    "HPI_+4",
    "UE_+4",
    "ACTLTV",
    "TDSR",
    "GDSR",
    "DSR_DIFF",
    "NUM_OF_BOR",
    "PRD_LOAN_TERM",
    "PRD_AMORTIZATION",
    "EFFECT_SID",
    "PRD_PORT",
    "PRD_FOREIGN_BORROWER",
    "PRK_PROPERTY_AGE",
    "PRK_PROPERTY_TYPE",
    "PRK_NUM_OF_UNITS",
    "PRK_OCCUPANCY",
    "PRK_CONS_TYPE",
    "PRK_TENURE",
    "RURAL_IND",
    "POD_MARKET_CODE",
    "POD_SOURCE_OF_BUSS_TYPE",
    "PRD_LOAN_PURPOSE",
    "BORYREMP",
    "LOB_RENAMED",
    "POD_OMNI_SCORE",
    "category",
    "BOOKYEAR",
    "STATUS",
    "target",
    "POD_INSURANCE_IN_FORCE",
    "RAS_OMNI",
    "lgd",
    "TOTAL_INCURRED_LOSS"
]

df_raw = pd.read_csv(
    "3_final_combined_dataset_all.csv",
    usecols=features
)

print(df_raw.shape)
display(df_raw.head())


# ============================================================
# 2. BASIC DATA CHECKS
# ============================================================

print("Shape:", df_raw.shape)

print("\nTarget distribution:")
display(df_raw["target"].value_counts(dropna=False))

print("\nStatus distribution:")
display(df_raw["STATUS"].value_counts(dropna=False))

print("\nBookyear range:")
display(df_raw["BOOKYEAR"].describe())

print("\nMissing rate top 20:")
display(
    df_raw.isna()
    .mean()
    .sort_values(ascending=False)
    .head(20)
    .to_frame("missing_rate")
)


# ============================================================
# 3. FILTER DATA
# ============================================================
# Keep funded + selected declined records.
# Remove Others.
# This follows your current logic.

df = df_raw.copy()

mask_declined = (
    (df["STATUS"] == "Declined") &
    (df["CRD_SCORE"] < 700) &
    (df["RAS_OMNI"] == "E") &
    (df["EFFECT_SID"].between(200900, 201712))
)

mask_keep = (
    (df["STATUS"] != "Declined") &
    (df["STATUS"] != "Others")
) | mask_declined

df_filtered = df.loc[mask_keep].copy()

# historical model-development window
df_main = df_filtered[df_filtered["BOOKYEAR"].between(2009, 2017)].copy()

# optional recent monitoring window
df_new = df_filtered[df_filtered["BOOKYEAR"].between(2023, 2025)].copy()

df = df_main.copy()

print("Filtered main data shape:", df.shape)
display(df.head())


# ============================================================
# 4. REMOVE LEAKAGE / POST-OUTCOME VARIABLES
# ============================================================
# IMPORTANT:
# These variables should NOT be used for feature-target relationship,
# t-tests, chi-square, IV, correlation, or model-feature assessment.
#
# Reason:
# category, lgd, and TOTAL_INCURRED_LOSS are outcome/post-event variables.
# They directly encode target behavior and cause leakage.

leakage_cols = [
    "category",
    "lgd",
    "TOTAL_INCURRED_LOSS"
]

id_cols = [
    "APPLICATION_NUMBER"
]

non_model_cols = [
    "target",
    "STATUS",
    "BOOKYEAR",
    "EFFECT_SID"
]

drop_from_analysis = leakage_cols + id_cols + non_model_cols

analysis_cols = [c for c in df.columns if c not in drop_from_analysis]

print("Columns excluded from analysis:")
print(drop_from_analysis)

print("\nNumber of analysis columns:", len(analysis_cols))


# ============================================================
# 5. CREATE ALL MODEL AND FUNDED MODEL DATASETS
# ============================================================

all_model = df[df["target"].notna()].copy()

funded_model = df[
    (df["target"].notna()) &
    (df["STATUS"] == "Funded")
].copy()

print("All model:", all_model.shape)
print("Funded model:", funded_model.shape)


# ============================================================
# 6. DATASET SUMMARY
# ============================================================

def dataset_summary(data, name):
    total = len(data)
    bads = (data["target"] == 1).sum()
    goods = (data["target"] == 0).sum()
    return pd.DataFrame({
        "model": [name],
        "rows": [total],
        "target_bad_count": [bads],
        "target_good_count": [goods],
        "bad_rate": [bads / total if total > 0 else np.nan],
        "missing_target": [data["target"].isna().sum()]
    })

summary = pd.concat([
    dataset_summary(all_model, "all_model"),
    dataset_summary(funded_model, "funded_model")
], ignore_index=True)

display(summary)


# ============================================================
# 7. TARGET DISTRIBUTION PLOTS
# ============================================================

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

for ax, data, title in zip(
    axes,
    [all_model, funded_model],
    ["All Model Target Distribution", "Funded Model Target Distribution"]
):
    plot_df = data["target"].value_counts().sort_index()
    bars = ax.bar(plot_df.index.astype(str), plot_df.values)

    for bar, value in zip(bars, plot_df.values):
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height(),
            f"{value:,.0f}",
            ha="center",
            va="bottom"
        )

    ax.set_title(title)
    ax.set_xlabel("Target")
    ax.set_ylabel("Count")

plt.tight_layout()
plt.show()


# ============================================================
# 8. IDENTIFY NUMERIC AND CATEGORICAL FEATURES
# ============================================================

numeric_cols = [
    c for c in analysis_cols
    if pd.api.types.is_numeric_dtype(df[c])
]

categorical_cols = [
    c for c in analysis_cols
    if c not in numeric_cols
]

# Treat selected numeric-coded variables as categorical
force_categorical = [
    "CRD_SCORE_bin",
    "NUM_OF_BOR",
    "PRK_NUM_OF_UNITS"
]

for c in force_categorical:
    if c in numeric_cols:
        numeric_cols.remove(c)
    if c in analysis_cols and c not in categorical_cols:
        categorical_cols.append(c)

print("Numeric columns:", numeric_cols)
print("\nCategorical columns:", categorical_cols)


# ============================================================
# 9. NUMERIC DISTRIBUTION SUMMARY
# ============================================================

def numeric_distribution(data, features, model_name):
    rows = []

    for col in features:
        s = data[col]

        rows.append({
            "model": model_name,
            "feature": col,
            "count": s.notna().sum(),
            "missing": s.isna().sum(),
            "missing_rate": s.isna().mean(),
            "mean": s.mean(),
            "std": s.std(),
            "min": s.min(),
            "p1": s.quantile(0.01),
            "p5": s.quantile(0.05),
            "p25": s.quantile(0.25),
            "median": s.quantile(0.50),
            "p75": s.quantile(0.75),
            "p95": s.quantile(0.95),
            "p99": s.quantile(0.99),
            "max": s.max()
        })

    return pd.DataFrame(rows)

num_dist = pd.concat([
    numeric_distribution(all_model, numeric_cols, "all_model"),
    numeric_distribution(funded_model, numeric_cols, "funded_model")
], ignore_index=True)

display(num_dist)


# ============================================================
# 10. NUMERIC DISTRIBUTION COMPARISON TABLE
# ============================================================

num_compare = (
    num_dist
    .pivot(index="feature", columns="model")
)

# flatten column names
num_compare.columns = [f"{metric}_{model}" for metric, model in num_compare.columns]
num_compare = num_compare.reset_index()

num_compare["mean_diff_all_minus_funded"] = (
    num_compare["mean_all_model"] - num_compare["mean_funded_model"]
)

num_compare["missing_rate_diff_all_minus_funded"] = (
    num_compare["missing_rate_all_model"] - num_compare["missing_rate_funded_model"]
)

display(num_compare.sort_values("mean_diff_all_minus_funded", key=abs, ascending=False))


# ============================================================
# 11. CATEGORICAL DISTRIBUTION SUMMARY
# ============================================================

def categorical_distribution(data, features, model_name, top_n=None):
    rows = []

    for col in features:
        temp = (
            data[col]
            .astype("object")
            .fillna("MISSING")
            .value_counts(dropna=False)
            .reset_index()
        )

        temp.columns = ["level", "count"]
        temp["feature"] = col
        temp["model"] = model_name
        temp["percent"] = temp["count"] / len(data)

        if top_n is not None:
            temp = temp.head(top_n)

        rows.append(temp)

    return pd.concat(rows, ignore_index=True)

cat_dist = pd.concat([
    categorical_distribution(all_model, categorical_cols, "all_model"),
    categorical_distribution(funded_model, categorical_cols, "funded_model")
], ignore_index=True)

display(cat_dist)


# ============================================================
# 12. CATEGORICAL FUNDED VS ALL COMPARISON
# ============================================================

cat_all = categorical_distribution(all_model, categorical_cols, "all_model")
cat_funded = categorical_distribution(funded_model, categorical_cols, "funded_model")

cat_compare = cat_all.merge(
    cat_funded,
    on=["feature", "level"],
    how="outer",
    suffixes=("_all", "_funded")
)

for col in ["count_all", "count_funded", "percent_all", "percent_funded"]:
    cat_compare[col] = cat_compare[col].fillna(0)

cat_compare["pct_diff_all_minus_funded"] = (
    cat_compare["percent_all"] - cat_compare["percent_funded"]
)

display(
    cat_compare
    .sort_values("pct_diff_all_minus_funded", key=abs, ascending=False)
    .head(50)
)


# ============================================================
# 13. PSI FUNCTION
# ============================================================

def calculate_psi_from_percents(expected_pct, actual_pct, eps=1e-6):
    expected_pct = np.where(expected_pct == 0, eps, expected_pct)
    actual_pct = np.where(actual_pct == 0, eps, actual_pct)

    return np.sum((actual_pct - expected_pct) * np.log(actual_pct / expected_pct))


def psi_numeric(expected, actual, bins=10):
    expected = expected.dropna()
    actual = actual.dropna()

    if expected.nunique() < 2 or actual.nunique() < 2:
        return np.nan

    try:
        breakpoints = np.unique(np.quantile(expected, np.linspace(0, 1, bins + 1)))
        if len(breakpoints) <= 2:
            return np.nan

        expected_bins = pd.cut(expected, bins=breakpoints, include_lowest=True)
        actual_bins = pd.cut(actual, bins=breakpoints, include_lowest=True)

        expected_pct = expected_bins.value_counts(normalize=True, sort=False).values
        actual_pct = actual_bins.value_counts(normalize=True, sort=False).reindex(
            expected_bins.value_counts(sort=False).index,
            fill_value=0
        ).values

        return calculate_psi_from_percents(expected_pct, actual_pct)

    except Exception:
        return np.nan


def psi_categorical(expected, actual):
    expected = expected.astype("object").fillna("MISSING")
    actual = actual.astype("object").fillna("MISSING")

    levels = sorted(set(expected.unique()).union(set(actual.unique())))

    expected_pct = expected.value_counts(normalize=True).reindex(levels, fill_value=0).values
    actual_pct = actual.value_counts(normalize=True).reindex(levels, fill_value=0).values

    return calculate_psi_from_percents(expected_pct, actual_pct)


def psi_interpretation(psi):
    if pd.isna(psi):
        return "Not calculated"
    elif psi < 0.10:
        return "Stable"
    elif psi < 0.25:
        return "Moderate shift"
    else:
        return "Significant shift"


# ============================================================
# 14. PSI: FUNDED VS ALL
# ============================================================

psi_rows = []

for col in numeric_cols:
    val = psi_numeric(all_model[col], funded_model[col])
    psi_rows.append({
        "feature": col,
        "type": "numeric",
        "psi": val,
        "interpretation": psi_interpretation(val)
    })

for col in categorical_cols:
    val = psi_categorical(all_model[col], funded_model[col])
    psi_rows.append({
        "feature": col,
        "type": "categorical",
        "psi": val,
        "interpretation": psi_interpretation(val)
    })

psi_table = pd.DataFrame(psi_rows).sort_values("psi", ascending=False)

display(psi_table)


# ============================================================
# 15. NUMERIC FEATURE VS TARGET RELATIONSHIP
# ============================================================

def numeric_target_relationship(data, features, model_name, bins=10, min_count=100):
    rows = []

    for col in features:
        temp = data[[col, "target"]].copy()
        temp = temp[temp["target"].notna()]
        temp = temp[temp[col].notna()]

        if temp[col].nunique() < 2 or len(temp) < min_count:
            continue

        try:
            temp["bin"] = pd.qcut(temp[col], q=bins, duplicates="drop")
        except Exception:
            continue

        grouped = (
            temp
            .groupby("bin", observed=False)
            .agg(
                count=("target", "size"),
                bad_count=("target", "sum"),
                bad_rate=("target", "mean"),
                min_value=(col, "min"),
                max_value=(col, "max"),
                mean_value=(col, "mean")
            )
            .reset_index()
        )

        grouped = grouped[grouped["count"] >= min_count].copy()
        grouped["feature"] = col
        grouped["model"] = model_name
        grouped["bad_rate_pct"] = grouped["bad_rate"] * 100

        rows.append(grouped)

    if len(rows) == 0:
        return pd.DataFrame()

    return pd.concat(rows, ignore_index=True)

num_target_rel = pd.concat([
    numeric_target_relationship(all_model, numeric_cols, "all_model"),
    numeric_target_relationship(funded_model, numeric_cols, "funded_model")
], ignore_index=True)

display(num_target_rel)


# ============================================================
# 16. CATEGORICAL FEATURE VS TARGET RELATIONSHIP
# ============================================================

def categorical_target_relationship(data, features, model_name, min_count=100):
    rows = []

    for col in features:
        temp = data[[col, "target"]].copy()
        temp = temp[temp["target"].notna()]
        temp[col] = temp[col].astype("object").fillna("MISSING")

        grouped = (
            temp
            .groupby(col)
            .agg(
                count=("target", "size"),
                bad_count=("target", "sum"),
                bad_rate=("target", "mean")
            )
            .reset_index()
        )

        grouped = grouped[grouped["count"] >= min_count].copy()
        grouped = grouped.rename(columns={col: "level"})
        grouped["feature"] = col
        grouped["model"] = model_name
        grouped["bad_rate_pct"] = grouped["bad_rate"] * 100

        rows.append(grouped)

    return pd.concat(rows, ignore_index=True)

cat_target_rel = pd.concat([
    categorical_target_relationship(all_model, categorical_cols, "all_model"),
    categorical_target_relationship(funded_model, categorical_cols, "funded_model")
], ignore_index=True)

display(cat_target_rel)


# ============================================================
# 17. MONOTONICITY CHECK FOR NUMERIC FEATURES
# ============================================================

def monotonicity_check(num_target_rel):
    rows = []

    for (model, feature), g in num_target_rel.groupby(["model", "feature"]):
        g = g.sort_values("min_value").copy()
        br = g["bad_rate"].values

        increasing = np.all(np.diff(br) >= 0)
        decreasing = np.all(np.diff(br) <= 0)

        rows.append({
            "model": model,
            "feature": feature,
            "bins": len(g),
            "min_bad_rate": np.min(br),
            "max_bad_rate": np.max(br),
            "bad_rate_range": np.max(br) - np.min(br),
            "monotonic_direction": (
                "Increasing" if increasing else
                "Decreasing" if decreasing else
                "Non-monotonic"
            )
        })

    return pd.DataFrame(rows)

mono_table = monotonicity_check(num_target_rel)

display(
    mono_table
    .sort_values(["model", "bad_rate_range"], ascending=[True, False])
)


# ============================================================
# 18. IV / WOE FUNCTIONS
# ============================================================

def calculate_iv_numeric(data, feature, target="target", bins=10, eps=1e-6):
    temp = data[[feature, target]].copy()
    temp = temp[temp[target].notna()]
    temp = temp[temp[feature].notna()]

    if temp[feature].nunique() < 2:
        return np.nan

    try:
        temp["bin"] = pd.qcut(temp[feature], q=bins, duplicates="drop")
    except Exception:
        return np.nan

    grouped = (
        temp
        .groupby("bin", observed=False)
        .agg(
            total=(target, "size"),
            bad=(target, "sum")
        )
        .reset_index()
    )

    grouped["good"] = grouped["total"] - grouped["bad"]

    total_bad = grouped["bad"].sum()
    total_good = grouped["good"].sum()

    if total_bad == 0 or total_good == 0:
        return np.nan

    grouped["bad_dist"] = grouped["bad"] / total_bad
    grouped["good_dist"] = grouped["good"] / total_good

    grouped["bad_dist"] = grouped["bad_dist"].replace(0, eps)
    grouped["good_dist"] = grouped["good_dist"].replace(0, eps)

    grouped["woe"] = np.log(grouped["good_dist"] / grouped["bad_dist"])
    grouped["iv_component"] = (grouped["good_dist"] - grouped["bad_dist"]) * grouped["woe"]

    return grouped["iv_component"].sum()


def calculate_iv_categorical(data, feature, target="target", min_count=100, eps=1e-6):
    temp = data[[feature, target]].copy()
    temp = temp[temp[target].notna()]
    temp[feature] = temp[feature].astype("object").fillna("MISSING")

    grouped = (
        temp
        .groupby(feature)
        .agg(
            total=(target, "size"),
            bad=(target, "sum")
        )
        .reset_index()
    )

    grouped = grouped[grouped["total"] >= min_count].copy()
    grouped["good"] = grouped["total"] - grouped["bad"]

    total_bad = grouped["bad"].sum()
    total_good = grouped["good"].sum()

    if total_bad == 0 or total_good == 0:
        return np.nan

    grouped["bad_dist"] = grouped["bad"] / total_bad
    grouped["good_dist"] = grouped["good"] / total_good

    grouped["bad_dist"] = grouped["bad_dist"].replace(0, eps)
    grouped["good_dist"] = grouped["good_dist"].replace(0, eps)

    grouped["woe"] = np.log(grouped["good_dist"] / grouped["bad_dist"])
    grouped["iv_component"] = (grouped["good_dist"] - grouped["bad_dist"]) * grouped["woe"]

    return grouped["iv_component"].sum()


def iv_strength(iv):
    if pd.isna(iv):
        return "Not calculated"
    elif iv < 0.02:
        return "Not predictive"
    elif iv < 0.10:
        return "Weak"
    elif iv < 0.30:
        return "Medium"
    elif iv < 0.50:
        return "Strong"
    else:
        return "Very strong / check leakage"


# ============================================================
# 19. IV TABLE
# ============================================================

iv_rows = []

for model_name, data in {
    "all_model": all_model,
    "funded_model": funded_model
}.items():

    for col in numeric_cols:
        val = calculate_iv_numeric(data, col)
        iv_rows.append({
            "model": model_name,
            "feature": col,
            "type": "numeric",
            "iv": val,
            "iv_strength": iv_strength(val)
        })

    for col in categorical_cols:
        val = calculate_iv_categorical(data, col)
        iv_rows.append({
            "model": model_name,
            "feature": col,
            "type": "categorical",
            "iv": val,
            "iv_strength": iv_strength(val)
        })

iv_table = pd.DataFrame(iv_rows).sort_values(["model", "iv"], ascending=[True, False])

display(iv_table)


# ============================================================
# 20. UNIVARIATE AUC FOR NUMERIC FEATURES
# ============================================================

auc_rows = []

for model_name, data in {
    "all_model": all_model,
    "funded_model": funded_model
}.items():

    for col in numeric_cols:
        temp = data[[col, "target"]].dropna()

        if temp["target"].nunique() < 2 or temp[col].nunique() < 2:
            continue

        try:
            auc = roc_auc_score(temp["target"], temp[col])
            auc_adj = max(auc, 1 - auc)

            auc_rows.append({
                "model": model_name,
                "feature": col,
                "auc": auc,
                "auc_direction_adjusted": auc_adj
            })
        except Exception:
            continue

auc_table = pd.DataFrame(auc_rows).sort_values(
    ["model", "auc_direction_adjusted"],
    ascending=[True, False]
)

display(auc_table)


# ============================================================
# 21. WELCH T-TEST + COHEN'S D
# ============================================================
# Keep this as supporting evidence only.
# Do NOT use p-values alone for feature selection.

def cohen_d(x0, x1):
    x0 = x0.dropna()
    x1 = x1.dropna()

    n0 = len(x0)
    n1 = len(x1)

    if n0 < 2 or n1 < 2:
        return np.nan

    pooled_std = np.sqrt(
        ((n0 - 1) * x0.var() + (n1 - 1) * x1.var()) /
        (n0 + n1 - 2)
    )

    if pooled_std == 0:
        return np.nan

    return (x1.mean() - x0.mean()) / pooled_std


def significance_level(p):
    if p < 0.001:
        return "*** Highly Significant"
    elif p < 0.01:
        return "** Very Significant"
    elif p < 0.05:
        return "* Significant"
    else:
        return "NS Not Significant"


def run_ttest(data, features):
    results = []

    for col in features:
        temp = data[[col, "target"]].replace([np.inf, -np.inf], np.nan).dropna()

        grp0 = temp[temp["target"] == 0][col]
        grp1 = temp[temp["target"] == 1][col]

        if len(grp0) < 2 or len(grp1) < 2:
            continue

        t_stat, p_value = ttest_ind(grp0, grp1, equal_var=False)

        results.append({
            "feature": col,
            "mean_target_0": grp0.mean(),
            "mean_target_1": grp1.mean(),
            "mean_diff_bad_minus_good": grp1.mean() - grp0.mean(),
            "t_statistic": t_stat,
            "p_value": p_value,
            "cohen_d": cohen_d(grp0, grp1),
            "target_0_count": len(grp0),
            "target_1_count": len(grp1),
            "significance_level": significance_level(p_value)
        })

    return pd.DataFrame(results).sort_values(
        "cohen_d",
        key=abs,
        ascending=False
    )

ttest_all = run_ttest(all_model, numeric_cols)
ttest_funded = run_ttest(funded_model, numeric_cols)

print("ALL MODEL T-TEST")
display(ttest_all)

print("FUNDED MODEL T-TEST")
display(ttest_funded)


# ============================================================
# 22. CHI-SQUARE + CRAMER'S V FOR CATEGORICAL FEATURES
# ============================================================
# Cramer's V is more useful than p-value because it measures strength.

def cramers_v(confusion_matrix):
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()

    if n == 0:
        return np.nan

    r, k = confusion_matrix.shape

    return np.sqrt(chi2 / (n * (min(k - 1, r - 1))))


def run_chi_square(data, features):
    results = []

    for col in features:
        temp = data[[col, "target"]].copy()
        temp = temp[temp["target"].notna()]
        temp[col] = temp[col].astype("object").fillna("MISSING")

        table = pd.crosstab(temp[col], temp["target"])

        if table.shape[0] < 2 or table.shape[1] < 2:
            continue

        try:
            chi2, p, dof, expected = chi2_contingency(table)
            cv = cramers_v(table)

            results.append({
                "feature": col,
                "chi_square": chi2,
                "p_value": p,
                "cramers_v": cv,
                "dof": dof,
                "significance_level": significance_level(p)
            })

        except Exception:
            continue

    return pd.DataFrame(results).sort_values(
        "cramers_v",
        ascending=False
    )

chi_all = run_chi_square(all_model, categorical_cols)
chi_funded = run_chi_square(funded_model, categorical_cols)

print("ALL MODEL CHI-SQUARE")
display(chi_all)

print("FUNDED MODEL CHI-SQUARE")
display(chi_funded)


# ============================================================
# 23. CORRELATION ANALYSIS
# ============================================================

all_corr = all_model[numeric_cols].corr()
funded_corr = funded_model[numeric_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(all_corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("All Model Correlation Heatmap")
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 10))
sns.heatmap(funded_corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Funded Model Correlation Heatmap")
plt.tight_layout()
plt.show()


def high_corr_pairs(corr_matrix, threshold=0.70):
    corr_abs = corr_matrix.abs()
    upper = corr_abs.where(np.triu(np.ones(corr_abs.shape), k=1).astype(bool))

    pairs = (
        upper
        .stack()
        .reset_index()
    )

    pairs.columns = ["feature_1", "feature_2", "abs_corr"]
    return pairs[pairs["abs_corr"] >= threshold].sort_values("abs_corr", ascending=False)

print("High correlation pairs - All Model")
display(high_corr_pairs(all_corr))

print("High correlation pairs - Funded Model")
display(high_corr_pairs(funded_corr))


# ============================================================
# 24. IMPORTANT NUMERIC BAD RATE PLOTS
# ============================================================

important_numeric = [
    "CRD_SCORE",
    "POD_OMNI_SCORE",
    "CRD_MONTHS_TU",
    "CRD_OPEN_TRADE_TU",
    "CRD_PUBLIC_DEROG_TU",
    "HPI_+4",
    "ACTLTV",
    "TDSR",
    "GDSR",
    "DSR_DIFF",
    "PRD_LOAN_TERM",
    "PRD_AMORTIZATION",
    "PRK_PROPERTY_AGE",
    "BORYREMP",
    "POD_INSURANCE_IN_FORCE"
]

important_numeric = [c for c in important_numeric if c in numeric_cols]

for col in important_numeric:
    plot_df = num_target_rel[num_target_rel["feature"] == col].copy()

    if plot_df.empty:
        continue

    plot_df = plot_df.sort_values(["model", "min_value"])
    plot_df["bin_label"] = (
        plot_df["min_value"].round(1).astype(str)
        + " - "
        + plot_df["max_value"].round(1).astype(str)
    )

    plt.figure(figsize=(12, 5))

    for model_name in ["all_model", "funded_model"]:
        temp = plot_df[plot_df["model"] == model_name].copy()

        if temp.empty:
            continue

        plt.plot(
            temp["bin_label"],
            temp["bad_rate_pct"],
            marker="o",
            label=model_name
        )

    plt.title(f"{col}: Bad Rate by Quantile Bin")
    plt.xlabel(col)
    plt.ylabel("Bad Rate (%)")
    plt.xticks(rotation=45, ha="right")
    plt.legend()
    plt.tight_layout()
    plt.show()


# ============================================================
# 25. IMPORTANT CATEGORICAL BAD RATE PLOTS
# ============================================================

important_categorical = [
    "CRD_SCORE_bin",
    "CRA_COLLECTION",
    "CRA_JUDGEMENT",
    "CRA_BANKRUPTCY_FORECLOSURE_IND",
    "ECON_REGION",
    "NUM_OF_BOR",
    "PRD_PORT",
    "PRD_FOREIGN_BORROWER",
    "PRK_PROPERTY_TYPE",
    "PRK_NUM_OF_UNITS",
    "PRK_OCCUPANCY",
    "PRK_CONS_TYPE",
    "PRK_TENURE",
    "RURAL_IND",
    "POD_MARKET_CODE",
    "POD_SOURCE_OF_BUSS_TYPE",
    "PRD_LOAN_PURPOSE",
    "LOB_RENAMED"
]

important_categorical = [c for c in important_categorical if c in categorical_cols]

for col in important_categorical:
    plot_df = cat_target_rel[cat_target_rel["feature"] == col].copy()

    if plot_df.empty:
        continue

    for model_name in ["all_model", "funded_model"]:
        temp = plot_df[plot_df["model"] == model_name].copy()

        if temp.empty:
            continue

        temp = temp.sort_values("bad_rate_pct", ascending=False).head(20)

        plt.figure(figsize=(12, 5))
        bars = plt.bar(temp["level"].astype(str), temp["bad_rate_pct"])

        for bar, rate, count in zip(bars, temp["bad_rate_pct"], temp["count"]):
            plt.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_height(),
                f"{rate:.1f}%\n(n={count:,.0f})",
                ha="center",
                va="bottom",
                fontsize=8
            )

        plt.title(f"{model_name}: {col} vs Target Bad Rate")
        plt.xlabel(col)
        plt.ylabel("Bad Rate (%)")
        plt.xticks(rotation=45, ha="right")
        plt.tight_layout()
        plt.show()


# ============================================================
# 26. CRD_SCORE_BIN SPECIAL PLOT
# ============================================================

def plot_score_bin_bad_rate(data, model_name, score_col="CRD_SCORE_bin"):
    temp = data[[score_col, "target"]].copy()
    temp[score_col] = temp[score_col].astype("object").fillna("MISSING")

    grouped = (
        temp
        .groupby(score_col)
        .agg(
            count=("target", "size"),
            bad_count=("target", "sum"),
            bad_rate=("target", "mean")
        )
        .reset_index()
    )

    grouped["bad_rate_pct"] = grouped["bad_rate"] * 100

    # custom sort: numeric score bins first, missing last
    def sort_key(x):
        try:
            return float(x)
        except Exception:
            return 999

    grouped["sort_key"] = grouped[score_col].apply(sort_key)
    grouped = grouped.sort_values("sort_key")

    plt.figure(figsize=(12, 5))
    bars = plt.bar(grouped[score_col].astype(str), grouped["bad_rate_pct"])

    for bar, rate, count in zip(bars, grouped["bad_rate_pct"], grouped["count"]):
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height(),
            f"{rate:.1f}%\n(n={count:,.0f})",
            ha="center",
            va="bottom",
            fontsize=9
        )

    plt.title(f"{model_name}: {score_col} vs Target Bad Rate")
    plt.xlabel(score_col)
    plt.ylabel("Bad Rate (%)")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

    return grouped

score_all = plot_score_bin_bad_rate(all_model, "all_model")
score_funded = plot_score_bin_bad_rate(funded_model, "funded_model")

display(score_all)
display(score_funded)


# ============================================================
# 27. MISSINGNESS VS TARGET
# ============================================================
# Important because missing values can be predictive in credit data.

def missingness_target_analysis(data, features, model_name):
    rows = []

    for col in features:
        temp = data[[col, "target"]].copy()
        temp = temp[temp["target"].notna()]
        temp["is_missing"] = temp[col].isna()

        grouped = (
            temp
            .groupby("is_missing")
            .agg(
                count=("target", "size"),
                bad_count=("target", "sum"),
                bad_rate=("target", "mean")
            )
            .reset_index()
        )

        grouped["feature"] = col
        grouped["model"] = model_name
        rows.append(grouped)

    return pd.concat(rows, ignore_index=True)

missing_target = pd.concat([
    missingness_target_analysis(all_model, analysis_cols, "all_model"),
    missingness_target_analysis(funded_model, analysis_cols, "funded_model")
], ignore_index=True)

display(
    missing_target
    .sort_values(["feature", "model", "is_missing"])
)


# ============================================================
# 28. FINAL COMBINED FEATURE SUMMARY
# ============================================================

# Combine PSI, IV, monotonicity, AUC, t-test/effect-size where available.

iv_wide = iv_table.pivot_table(
    index="feature",
    columns="model",
    values="iv",
    aggfunc="first"
).reset_index()

iv_wide.columns = [
    "feature",
    "iv_all_model",
    "iv_funded_model"
]

psi_small = psi_table[["feature", "type", "psi", "interpretation"]].copy()

mono_wide = mono_table.pivot_table(
    index="feature",
    columns="model",
    values="monotonic_direction",
    aggfunc="first"
).reset_index()

mono_wide.columns = [
    "feature",
    "monotonic_all_model",
    "monotonic_funded_model"
]

auc_wide = auc_table.pivot_table(
    index="feature",
    columns="model",
    values="auc_direction_adjusted",
    aggfunc="first"
).reset_index()

auc_wide.columns = [
    "feature",
    "auc_all_model",
    "auc_funded_model"
]

ttest_all_small = ttest_all[["feature", "cohen_d", "mean_diff_bad_minus_good"]].copy()
ttest_all_small = ttest_all_small.rename(columns={
    "cohen_d": "cohen_d_all_model",
    "mean_diff_bad_minus_good": "mean_diff_bad_minus_good_all_model"
})

ttest_funded_small = ttest_funded[["feature", "cohen_d", "mean_diff_bad_minus_good"]].copy()
ttest_funded_small = ttest_funded_small.rename(columns={
    "cohen_d": "cohen_d_funded_model",
    "mean_diff_bad_minus_good": "mean_diff_bad_minus_good_funded_model"
})

final_feature_summary = (
    psi_small
    .merge(iv_wide, on="feature", how="left")
    .merge(mono_wide, on="feature", how="left")
    .merge(auc_wide, on="feature", how="left")
    .merge(ttest_all_small, on="feature", how="left")
    .merge(ttest_funded_small, on="feature", how="left")
)

display(
    final_feature_summary
    .sort_values("iv_all_model", ascending=False)
)


# ============================================================
# 29. EXPORT OUTPUTS
# ============================================================

with pd.ExcelWriter("funded_vs_all_model_analysis_outputs.xlsx") as writer:
    summary.to_excel(writer, sheet_name="dataset_summary", index=False)
    num_dist.to_excel(writer, sheet_name="numeric_distribution", index=False)
    num_compare.to_excel(writer, sheet_name="numeric_compare", index=False)
    cat_compare.to_excel(writer, sheet_name="categorical_compare", index=False)
    psi_table.to_excel(writer, sheet_name="psi", index=False)
    num_target_rel.to_excel(writer, sheet_name="numeric_target_rel", index=False)
    cat_target_rel.to_excel(writer, sheet_name="categorical_target_rel", index=False)
    mono_table.to_excel(writer, sheet_name="monotonicity", index=False)
    iv_table.to_excel(writer, sheet_name="iv", index=False)
    auc_table.to_excel(writer, sheet_name="auc", index=False)
    ttest_all.to_excel(writer, sheet_name="ttest_all", index=False)
    ttest_funded.to_excel(writer, sheet_name="ttest_funded", index=False)
    chi_all.to_excel(writer, sheet_name="chi_all", index=False)
    chi_funded.to_excel(writer, sheet_name="chi_funded", index=False)
    missing_target.to_excel(writer, sheet_name="missingness_target", index=False)
    final_feature_summary.to_excel(writer, sheet_name="final_feature_summary", index=False)

print("Export completed: funded_vs_all_model_analysis_outputs.xlsx")
```
