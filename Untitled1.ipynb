```python
def plot_stacked_numeric_distributions(
    all_model,
    funded_model,
    numeric_cols,
    bins=50
):
    for col in numeric_cols:
        plot_data = pd.concat([
            all_model[[col]].assign(model="all_model"),
            funded_model[[col]].assign(model="funded_model")
        ], ignore_index=True)

        plt.figure(figsize=(10, 5))

        sns.histplot(
            data=plot_data,
            x=col,
            hue="model",
            bins=bins,
            kde=True,
            stat="density",
            common_norm=False,
            element="step"
        )

        plt.title(f"All Model vs Funded Model - {col}")
        plt.xlabel(col)
        plt.ylabel("Density")
        plt.tight_layout()
        plt.show()
```

```python
def get_correlation_matrix(data, numeric_cols):
    return data[numeric_cols].corr()


def get_high_corr_pairs(corr_matrix, threshold=0.70):
    corr_abs = corr_matrix.abs()

    upper = corr_abs.where(
        np.triu(np.ones(corr_abs.shape), k=1).astype(bool)
    )

    pairs = upper.stack().reset_index()
    pairs.columns = ["feature_1", "feature_2", "abs_corr"]

    return (
        pairs[pairs["abs_corr"] >= threshold]
        .sort_values("abs_corr", ascending=False)
        .reset_index(drop=True)
    )


def plot_corr_heatmap(corr_matrix, title="Correlation Heatmap"):
    plt.figure(figsize=(14, 10))

    sns.heatmap(
        corr_matrix,
        annot=True,
        fmt=".2f",
        cmap="coolwarm",
        center=0
    )

    plt.title(title)
    plt.tight_layout()
    plt.show()


def run_corr_analysis(data, numeric_cols, model_name, threshold=0.70):
    corr_matrix = get_correlation_matrix(data, numeric_cols)

    plot_corr_heatmap(
        corr_matrix,
        title=f"{model_name} Correlation Heatmap"
    )

    high_corr = get_high_corr_pairs(
        corr_matrix,
        threshold=threshold
    )

    print(f"High Correlation Pairs - {model_name}")
    display(high_corr)

    return corr_matrix, high_corr
```
