```python
def plot_bad_rate_common_bins(
    data_1,
    data_2,
    feature,
    target_col="target",
    data_1_name="all_model",
    data_2_name="funded_model",
    bins=10
):

    base = data_1[feature].dropna()

    edges = np.unique(
        np.quantile(
            base,
            np.linspace(0, 1, bins + 1)
        )
    )

    if len(edges) <= 2:
        print(f"Skipping {feature} - insufficient unique values")
        return None

    rows = []

    for model_name, data in {
        data_1_name: data_1,
        data_2_name: data_2
    }.items():

        temp = data[[feature, target_col]].dropna().copy()

        temp["bin"] = pd.cut(
            temp[feature],
            bins=edges,
            include_lowest=True
        )

        grouped = (
            temp
            .groupby("bin", observed=False)
            .agg(
                count=(target_col, "size"),
                bad_count=(target_col, "sum"),
                bad_rate=(target_col, "mean")
            )
            .reset_index()
        )

        grouped["model"] = model_name
        grouped["bad_rate_pct"] = grouped["bad_rate"] * 100

        rows.append(grouped)

    plot_df = pd.concat(rows, ignore_index=True)

    plot_df["bin_label"] = plot_df["bin"].astype(str)

    plt.figure(figsize=(12, 5))

    for model_name in [data_1_name, data_2_name]:

        temp = plot_df[
            plot_df["model"] == model_name
        ]

        plt.plot(
            temp["bin_label"],
            temp["bad_rate_pct"],
            marker="o",
            linewidth=2,
            label=model_name
        )

    plt.title(f"{feature}: Bad Rate by Common Quantile Bin")
    plt.xlabel(feature)
    plt.ylabel("Bad Rate (%)")
    plt.xticks(rotation=45, ha="right")
    plt.legend()
    plt.tight_layout()
    plt.show()

    return plot_df
```

```python
for col in important_numeric:

    plot_bad_rate_common_bins(
        data_1=all_model,
        data_2=funded_model,
        feature=col,
        bins=10
    )
```

```python
def categorical_distribution(data, features):
    rows = []

    for col in features:

        temp = (
            data[col]
            .astype("object")
            .fillna("MISSING")
            .value_counts(dropna=False)
            .reset_index()
        )

        temp.columns = ["level", "count"]
        temp["feature"] = col
        temp["percent"] = temp["count"] / len(data)

        rows.append(temp)

    return pd.concat(rows, ignore_index=True)
```

```python
all_dist = categorical_distribution(
    all_model,
    categorical_cols
).rename(
    columns={
        "count": "all_model_count",
        "percent": "all_model_pct"
    }
)

funded_dist = categorical_distribution(
    funded_model,
    categorical_cols
).rename(
    columns={
        "count": "funded_model_count",
        "percent": "funded_model_pct"
    }
)
```

```python
cat_compare = (
    all_dist.merge(
        funded_dist,
        on=["feature", "level"],
        how="outer"
    )
    .fillna(0)
)
```
