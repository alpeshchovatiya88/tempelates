```python
def create_bad_rate_bins(
    data_1,
    data_2,
    feature,
    target_col="target",
    data_1_name="all_model",
    data_2_name="funded_model",
    bins=10
):
    base = data_1[feature].dropna()

    edges = np.unique(
        np.quantile(base, np.linspace(0, 1, bins + 1))
    )

    if len(edges) <= 2:
        return None

    rows = []

    for model_name, data in {
        data_1_name: data_1,
        data_2_name: data_2
    }.items():

        temp = data[[feature, target_col]].dropna().copy()

        temp["bin"] = pd.cut(
            temp[feature],
            bins=edges,
            include_lowest=True
        )

        grouped = (
            temp
            .groupby("bin", observed=False)
            .agg(
                count=(target_col, "size"),
                bad_count=(target_col, "sum"),
                bad_rate=(target_col, "mean")
            )
            .reset_index()
        )

        grouped["model"] = model_name
        grouped["bad_rate_pct"] = grouped["bad_rate"] * 100

        rows.append(grouped)

    return pd.concat(rows, ignore_index=True)
```

```python
def plot_bad_rate_stacked(
    plot_df,
    feature,
    model_col="model"
):

    temp = plot_df.copy()
    temp["bin_label"] = temp["bin"].astype(str)

    pivot_df = (
        temp.pivot(
            index="bin_label",
            columns=model_col,
            values="bad_rate_pct"
        )
        .fillna(0)
    )

    pivot_df.plot(
        kind="bar",
        stacked=True,
        figsize=(12, 5)
    )

    plt.title(f"{feature} - Bad Rate Comparison")
    plt.xlabel(feature)
    plt.ylabel("Bad Rate (%)")
    plt.xticks(rotation=45, ha="right")
    plt.legend(title="Dataset")
    plt.tight_layout()
    plt.show()
```

```python
for col in important_numeric:

    plot_df = create_bad_rate_bins(
        data_1=all_model,
        data_2=funded_model,
        feature=col,
        bins=10
    )

    if plot_df is None or plot_df.empty:
        continue

    plot_bad_rate_stacked(
        plot_df=plot_df,
        feature=col
    )
```
